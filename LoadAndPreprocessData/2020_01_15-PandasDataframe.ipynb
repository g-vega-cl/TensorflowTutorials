````
from __future__ import absolute_import, division, print_function, unicode_literals

# TensorFlow and tf.keras
try:
  # %tensorflow_version only exists in Colab.
  %tensorflow_version 2.x
except Exception:
  pass
import tensorflow as tf
from tensorflow import keras
import numpy as np
````

````
import pandas as pd
````

````
csv_file = tf.keras.utils.get_file('heart.csv', 'https://storage.googleapis.com/applied-dl/heart.csv')
````

````
csv_file #This is the file, but it has to be read
df = pd.read_csv(csv_file)
````

````
df #We get a pandas dataframe
````

````
df.dtypes #Check the types of data
````

````
df['thal'] = pd.Categorical(df['thal'])
df['thal'] = df.thal.cat.codes #Transform thal into integers. ISSUE: is this not bad? should we use one-hot insted?
#because integers like this mean a relation between the numbers, not a category?
````

````
df
````

````
#Load data using tf.data.Dataset #I should read the documentation:  https://www.tensorflow.org/api_docs/python/tf/data/Dataset
#Use tf.data.Dataset.from_tensor_slices to read the values from a pandas dataframe. 
#NOTE: after tutorials, there are guides, eg. https://www.tensorflow.org/guide/data   <- data guide. Do those next.
target = df.pop('target') #NOTE: cannot run this cell twice because the df now does not have the 'target column'
dataset = tf.data.Dataset.from_tensor_slices((df.values, target.values)) #This places the labels and the values together
````

````
dataset #Because we have not batched it, it only has the 13, as shape
````

````
for feat, targ in dataset.take(5): #Because dataset has 'two sets?' we can use the for loop
  print ('Features: {}, Target: {}'.format(feat, targ)) #we take 5 of the features (5 rows)
````

````
#NOTE: Since a pd.Series implements the __array__ protocol it can be used transparently nearly anywhere you would use a np.array or a tf.Tensor.
tf.constant(df['thal']) #Take every value? of the column thal
````

````
#Shuffle and batch the dataset
train_dataset = dataset.shuffle(len(df)).batch(1)
````

#Steps to make a dataset:

1. use:
```
dataset = tf.data.Dataset.from_tensor_slices((df.values, target.values))#This places the labels and the values together
```
to build the dataset with the values and the target:

2. use:
```
train_dataset = dataset.shuffle(len(df)).batch(1) #Shuffle and batch the dataset
```
to make the dataset (shuffled) in batches to be able to pass it properly to the model






````
(train_dataset) #This is a dataset ..that has batches?
````

````
#Create and train model
def get_compiled_model(): #Building a function to get the model already copiled
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
  ])

  model.compile(optimizer='adam',
                loss='binary_crossentropy',
                metrics=['accuracy'])
  return model
````

````
model = get_compiled_model()
model.fit(train_dataset, epochs=15)
model.summary()
````

````
predictions = model.predict(train_dataset) #Aqui te da un array con todas las predicciones #las 303
predictions.shape #las predicciones son de 0 a 1 parece, aunque aquí la max es .4
````

#Alternative to feature columns

Passing a dictionary as an input to a model is as easy as creating a matching dictionary of tf.keras.layers.Input layers, applying any pre-processing and stacking them up using the functional api. You can use this as an alternative to feature columns.

````
#Basically doing the same but using a dictionary as column (feature) names?
inputs = {key: tf.keras.layers.Input(shape=(), name=key) for key in df.keys()}
x = tf.stack(list(inputs.values()), axis=-1)

x = tf.keras.layers.Dense(10, activation='relu')(x)
output = tf.keras.layers.Dense(1, activation='sigmoid')(x)

model_func = tf.keras.Model(inputs=inputs, outputs=output)

model_func.compile(optimizer='adam',
                   loss='binary_crossentropy',
                   metrics=['accuracy'])
````

The easiest way to preserve the column structure of a pd.DataFrame when used with tf.data is to convert the pd.DataFrame to a dict, and slice that dictionary.

````
dict_slices = tf.data.Dataset.from_tensor_slices((df.to_dict('list'), target.values)).batch(16) #This is a batch dataset
for dict_slice in dict_slices.take(1): #take one batch, of 16, it has 16 'rows' of the attributes
  print (dict_slice)
````

````
model_func.fit(dict_slices, epochs=15)
#Same fitting, just with this new way of building the data
````
